```python
import requests
import matplotlib.pyplot as plt
import pandas as pd

API_KEY= 'F7Ju2SIo2qHfBKEVKXcnxUcbT9zDIckVSbLkudiV'

response = requests.get('https://developer.nps.gov/api/v1/parks?stateCode=md&api_key=' + API_KEY)

json = response.json()

topicsDict = {}

for item in json['data']:
    for topic in item['topics']:
        topicName = topic['name']
        if topicName in topicsDict.keys():
            topicsDict[topicName] +=1
        else:
            topicsDict[topicName] =1
```

```python
topics = list(topicsDict.keys())
values = list(topicsDict.values())

f, ax = plt.subplots(figsize=(25,7))
plt.bar(range(len(topicsDict)), values, tick_label=topics, color = 'red', width=0.8)
plt.xticks(rotation = 90)
plt.title('MD State Park Topics By Frequency', fontsize = 20)
plt.xlabel('Topics', fontsize = 18)
plt.ylabel('Frequency', fontsize=18)
plt.show()
```

```python
df = pd.DataFrame({'Topics': topics, 'Frequency': values})
	
df_sorted= df.sort_values('Frequency', ascending=False)

plt.figure(figsize=(25,10))
plt.bar('Topics', 'Frequency',data=df_sorted, color = 'red', width=0.8)
plt.xticks(rotation = 90)
plt.title('MD State Park Topics By Frequency', fontsize = 20)
plt.xlabel('Topics', fontsize = 18)
plt.ylabel('Frequency', fontsize=18)
plt.show()
```
