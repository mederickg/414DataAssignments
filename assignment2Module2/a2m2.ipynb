```python
import pandas as pd
from pathlib import Path
import numpy as np
import networkx as nx
```

```python
g = nx.Graph() # Build the graph
```

```python
path = 'data'
files = Path(path).glob('*.csv')

dfs = list()
for f in files:
    data = pd.read_csv(f)
    # .stem is method for pathlib objects to get the filename w/o the extension
    data['file'] = f.stem
    dfs.append(data)

df = pd.concat(dfs, ignore_index=True)

df
```

```python
for row in df.iterrows():
    home = row[1]['HomeTeam']
    g.add_node(home)

    away = row[1]['AwayTeam']
    g.add_node(away)
    
    g.add_edge(home, away)
```

```python
print("Nodes:", len(g.nodes))
print(g.nodes)
```

```python
nx.ego_graph(g, n='Man United', radius = 4)
```

```python
sorted(g.degree, key=lambda x: x[1], reverse=True)[:10]
```

```python
nx.write_graphml(g, "premLeague.graphml")
```
